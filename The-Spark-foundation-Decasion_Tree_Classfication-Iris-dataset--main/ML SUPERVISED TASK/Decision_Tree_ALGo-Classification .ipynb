# Label Encoder

In machine learning, we usually deal with datasets which contains multiple labels in one or more than one columns. These labels can be in the form of words or numbers. Label Encoding refers to converting the labels into numeric form so as to convert it into the machine-readable form

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Species'] = le.fit_transform(df['Species'])
df.head(10)
```

## Model Training
      
      
   ### Spliting Data 80 & 20  ratio 

```python
from sklearn.model_selection import train_test_split
# train - 80
# test - 20
X = df.drop(columns=['Species'])
Y = df['Species']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

## LogisticRegression

**Hear we use different Machine learning algorithms And we are checking 
  accuracy with the all model how much percentage it gives a accuracy**

```python
# logistic regression 
from sklearn.linear_model import LogisticRegression
Log_model = LogisticRegression()
```

```python
# model training
Log_model.fit(x_train, y_train)
```

```python
# print metric to get performance
print("Accuracy: ",Log_model.score(x_test, y_test) * 100)
```

### knn -k-nearest neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(x_train, y_train)
# print metric to get performance
print("Accuracy: ",model.score(x_test, y_test) * 100)
```

### Decision Tree

### Perform Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(x_train)
X_test_sc = sc.transform(x_test)
```

### Model_Building

### Decision Tree

```python
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.tree import DecisionTreeClassifier
Dt_model = DecisionTreeClassifier(criterion= 'entropy', random_state=0)
Dt_model.fit(x_train, y_train)
y_pred = Dt_model.predict(x_test)
# print metric to get performance
print("Accuracy: ",model.score(x_test, y_test) * 100)
```
